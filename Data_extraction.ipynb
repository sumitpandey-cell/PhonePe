```python
import pandas as pd
import os
from pathlib import Path
import json
```

```python
BASE_PATH = Path("C:/Users/sumit/OneDrive/Documents/Data_analytics_project/PhonePe/data")
def load_json_files(base_dir):
    json_files = []
    for root,files in os.walk(base_dir):
        for file in files:
            if file.endswith('.json'):
                json_files.append(os.path.join(root, file))
    return json_files
```

```python
# Aggregate user data
agg_tr_file = []

for state in (BASE_PATH / "aggregated" / "user" / "country" / "india" / "state").iterdir():
    for year in state.iterdir():
        for quarter in year.iterdir():
            with open(quarter, 'r') as f:
                data = json.load(f)
                userData = data.get('data', {}).get('usersByDevice')
                if userData:
                    for entry in userData:
                        agg_tr_file.append({
                            "State": state.name,
                            "Year": str(year.name),
                            "Quarter": quarter.stem,
                            "Brand": entry.get('brand', 'Unknown'),
                            "Count": entry.get('count', 0),
                            "Percentage": entry.get('percentage', 0)
                        })
                else:
                    agg_tr_file.append({
                        "State": state.name,
                        "Year": str(year.name),
                        "Quarter": quarter.stem,
                        "Brand": None,
                        "Count": None,
                        "Percentage": None
                    })
                
df_agg_user = pd.DataFrame(agg_tr_file)
df_agg_user.head()
```

# Preprocessing

```python
def clean_data(df, number_cols, string_cols):
    #remove duplicates
    df = df.copy()
    df = df.drop_duplicates()

    #remove whitespace from string columns with replace space the _
    for col in string_cols:
        df[col] = df[col].str.strip().str.replace(" ", "_") 
    # fill missing values in number columns with 0
    if number_cols:
        for col in number_cols:
            if col in df.columns:
                df[col] = df[col].fillna(0)

    # fill missing values in string columns with "Unknown"
    if string_cols:
        for col in string_cols:
            if col in df.columns:
                df[col] = df[col].fillna("Unknown")
                
    # Removing invalid values (negative amount/count) 
    if number_cols:  
        for col in number_cols:
            if col in df.columns:
                df[col] = df[col].apply(lambda x: x if x >= 0 else 0)


    return df
```

```python
# clean all dataframs
df_agg_txn = clean_data(df_agg_txn, ["Transaction_count", "Total_amount"], ["State","Year","Quarter","Transaction_type"])
df_agg_user = clean_data(df_agg_user, ["Count","Percentage"], ["State", "Brand","Year","Quarter"])
df_map_txn = clean_data(df_map_txn, ["Transaction_count", "Total_amount"], ["State", "District", "Year","Quarter"])
df_map_user = clean_data(df_map_user, ["App_opens", "Registered_users"], ["State", "District", "Year","Quarter"])
df_top_txn_dis = clean_data(df_top_txn_dis, ["Transaction_count", "Total_amount"], ["State", "District", "Year","Quarter"])
df_top_txn_pin = clean_data(df_top_txn_pin, ["Transaction_count", "Total_amount"], ["State", "Pincode", "Year","Quarter"])
df_top_user_dis = clean_data(df_top_user_dis, ["Register_users"], ["State", "District", "Year","Quarter"])
df_top_user_pin = clean_data(df_top_user_pin, ["Register_users"], ["State", "Pincode"])
```
